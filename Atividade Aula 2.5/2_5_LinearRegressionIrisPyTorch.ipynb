# Regressão Linear com PyTorch

## Objetivo

Este exemplo utiliza regressão linear para estimar o comprimento das sépalas da íris a partir do seu comprimento das pétalas.
Utiliza-se
- o cálculo do gradiente automático do PyTorch, porém
- a função de perda e
- atualização dos parâmetros é feita de forma explícita.
- A rede é criada com uma camada nn.Linear()

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinear_Otimizacao.png)

## Importação dos pacotes

```python
%matplotlib inline
import matplotlib.pyplot as plt

import numpy as np
from sklearn.datasets import load_iris

import torch
from torch import nn, optim
from torch.autograd import Variable

torch.manual_seed(1234)
```

## Leitura dos dados

```python
iris = load_iris()
data = iris.data[iris.target==1,::2]  # comprimento das sépalas e pétalas, indices 0 e 2

x_train = data[:,0:1].astype(np.float32)
y_train = data[:,1:2].astype(np.float32)

n_samples = x_train.shape[0]
print('x_train.shape:',x_train.shape, x_train.dtype)
print('y_train.shape:',y_train.shape, y_train.dtype)

print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

### Normalização dos dados

```python
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()
```

```python
print('x_train[:5]:\n', x_train[:5])
print('y_train[:5]:\n', y_train[:5])
```

```python
x_train_bias = np.hstack([np.ones(shape=(n_samples,1)), x_train])
```

```python
x_train_bias
```

```python
print('x_train_bias.shape:',x_train_bias.shape, x_train_bias.dtype)
```

```python
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)
print(y_train)
```

## Model

![alt text](https://raw.githubusercontent.com/vcasadei/images/master/RegressaoLinearNeuronio.png)

```python
#Inicialização usando toch.nn.Linear

model = torch.nn.Linear(in_features=2, out_features=1, bias=False)
```

### Verificando a inicialização dos parâmetros

```python
model.weight.data = torch.zeros(1,2)
torch.nn.init.uniform(model.weight.data, -0.1, 0.1)
model.weight
```

### Testando o predict da rede

```python
xx = torch.ones((5,2))
xx
```

```python
model(Variable(xx))
```

## Treinamento

```python
num_epochs = 100
learning_rate = 0.5
for epoch in range(num_epochs):
    #inputs = Variable(x_train_bias)
    #target = Variable(y_train)
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    #loss cálculo da função de perda
    loss = torch.mean((y_pred - target)**2)

    # backward propagation - cálculo gradient
    loss.backward()

    # gradiente descendente
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # zeragem parâmetros próximo cálculo do gradiente
    model.weight.grad.data.zero_()

    # verbose
    if (epoch+1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch+1, num_epochs, loss.data))
```

## Avaliação

```python
y_pred = model(Variable(x_train_bias))
plt.plot(x_train, y_train.numpy(), 'ro', label='Original data')
plt.plot(x_train, y_pred.data.numpy(), 'kx-', label='Fitting Line')
plt.show()
```

- 3) Crie duas listas `w0_list` e `w1_list` para guardar a evolução dos valores dos
   parâmetros durante o laço de treinamento.
   Modifique o exemplo do gráfico tipo **scatter plot** para visualizar a variação
   dos parâmetros do modelo duas dimensões.
   Marque neste gráfico o valor ótimo do parâmetro dado pela solução

```python
# Consolidado

import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris

import torch
from torch import nn
from torch.autograd import Variable


torch.manual_seed(1234)

# -----------------------------
# 1) Leitura e preparação dos dados
# -----------------------------
iris = load_iris()
# Usando apenas a classe 1 (versicolor) e as colunas 0 e 2 (sépala, pétala)
data = iris.data[iris.target == 1, ::2]

# x = comprimento das sépalas, y = comprimento das pétalas
x_train = data[:, 0:1].astype(np.float32)
y_train = data[:, 1:2].astype(np.float32)

n_samples = x_train.shape[0]

# Normalização simples [0,1]
x_train -= x_train.min()
x_train /= x_train.max()
y_train -= y_train.min()
y_train /= y_train.max()

# Criação do x_train_bias: primeira coluna = 1s, segunda = x_train
x_train_bias = np.hstack([np.ones(shape=(n_samples, 1), dtype=np.float32), x_train])

# Converte tudo para tensores do PyTorch
x_train_bias = torch.FloatTensor(x_train_bias)
y_train      = torch.FloatTensor(y_train)


# -----------------------------
# 2) Modelo e inicialização
# -----------------------------
# Modelo linear sem bias
model = nn.Linear(in_features=2, out_features=1, bias=False)

# Inicializa pesos com valores pequenos aleatórios
nn.init.uniform_(model.weight.data, -0.1, 0.1)

print("Pesos iniciais:", model.weight.data)


# -----------------------------
# 3) Listas para guardar a evolução dos parâmetros (w0, w1)
# -----------------------------
w0_list = []
w1_list = []


# -----------------------------
# 4) Laço de treinamento com registro dos pesos
# -----------------------------
num_epochs = 100
learning_rate = 0.5

for epoch in range(num_epochs):
    inputs = x_train_bias
    target = y_train

    # forward - predict
    y_pred = model(inputs)

    # cálculo da função de perda (MSE)
    loss = torch.mean((y_pred - target) ** 2)

    # backward propagation - cálculo do gradiente
    loss.backward()

    # gradiente descendente (atualização manual dos pesos)
    model.weight.data = model.weight.data - learning_rate * model.weight.grad.data

    # guarda os valores atuais dos parâmetros após a atualização
    w0_list.append(model.weight.data[0, 0].item())
    w1_list.append(model.weight.data[0, 1].item())

    # zera os gradientes para a próxima iteração
    model.weight.grad.data.zero_()

    # imprime a loss a cada 20 épocas
    if (epoch + 1) % 20 == 0:
        print('Epoch[{}/{}], loss: {:.6f}'
              .format(epoch + 1, num_epochs, loss.item()))

print("Pesos finais (treinados):", model.weight.data)


# -----------------------------
# 5) Solução analítica (w_opt)
# -----------------------------
x_bias = x_train_bias              # [50, 2]
y      = y_train                   # [50, 1]
x_bias_t = torch.t(x_bias)         # [2, 50]

# Fórmula dos mínimos quadrados: w_opt = (XᵀX)^(-1) Xᵀ y
w_opt = (torch.inverse(x_bias_t.mm(x_bias)).mm(x_bias_t)).mm(y)   # [2, 1]

print("Parâmetros ótimos analíticos (w_opt):", w_opt.t())


# -----------------------------
# 6) Gráfico scatter da trajetória de (w0, w1) + ponto ótimo analítico
# -----------------------------
plt.figure()
plt.scatter(w0_list, w1_list, label='trajetória dos pesos')

# desenha setas indicando o caminho seguido pelos pesos no plano (w0, w1)
w0_old, w1_old = None, None
for (w0, w1) in zip(w0_list, w1_list):
    if w0_old is not None:
        plt.arrow(w0_old, w1_old, w0 - w0_old, w1 - w1_old,
                  head_length=0.01, head_width=0.01, shape='full',
                  length_includes_head=True)
    w0_old, w1_old = w0, w1

# marca no gráfico o valor ótimo dos parâmetros obtido pela solução analítica
w0_opt = w_opt[0].item()
w1_opt = w_opt[1].item()
plt.scatter([w0_opt], [w1_opt], marker='x', s=100, label='ótimo analítico')

plt.xlabel('w0')
plt.ylabel('w1')
plt.title('Evolução dos parâmetros w0 e w1 durante o treinamento')
plt.legend()
plt.grid(True)
plt.show()
```
